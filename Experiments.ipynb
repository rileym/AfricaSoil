```python
%matplotlib inline

import pandas as pd
import numpy as np
import sklearn.decomposition as decomp
import sklearn.preprocessing as pp
import sklearn.feature_selection as fs
import sklearn.pipeline as pipe
import sklearn.metrics as metrics
import matplotlib.pyplot as plt
import soilCleanTools as clean

# Read in data
train_df = pd.read_csv('./Data/training.csv', index_col='PIDN')

# Clean and split for readability
train_data_dict = clean.clean_data(train_df)
wave, spatial, targets = train_data_dict['wave'], train_data_dict['spatial'], train_data_dict['targets']
```

```python
steps = [('Correlation Filter', fs.SelectPercentile( score_func = fs.f_regression, percentile = 50)),
         ('PCA', decomp.PCA(n_components = 2)),
         ('Feature Standardization',pp.StandardScaler())]
superPCSclr = pipe.Pipeline(steps = steps)
Xtrans = pd.DataFrame(superPCSclr.fit_transform(X = wave, y = targets['P']), index = wave.index, columns = ['x','y'])
```
