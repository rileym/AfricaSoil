### Tranform skewed target variables

```python
targets['log_Ca'] = targets['Ca'].map(np.log)
targets['log_P'] = targets['P'].map(np.log)
targets['log_SOC'] = targets['SOC'].map(np.log)
targets.hist(bins=40)
```

```python
(fig, axes) = plt.subplots(nrows=3, ncols=3,sharex=True,sharey=True)
for (i, (column, target)) in enumerate(targets.iteritems()):
    spacial_df[spacial_df.columns-['Depth']].corrwith(target,axis=0).plot(kind='barh',xlim=[-1,1],ax=axes.flatten()[i],title=column)
```

```python
fig = plt.figure()
for (column, target) in targets.iteritems():
    if 'log' in column:
        wave_df.corrwith(target, axis=0).plot(ylim=[-1,1],label=column)

plt.gca().legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3, ncol=5, mode="expand", borderaxespad=0.)
```
